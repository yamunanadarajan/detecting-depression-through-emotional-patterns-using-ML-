```python
import gensim
import pandas as pd
```

```python
def readn(r,c):
  global em
  df = pd.read_fwf(r,lines=True)
  sh=df.shape
  print(df.columns)
  df[c] = df[c].astype(str)
  em = df[c].apply(gensim.utils.simple_preprocess)
  row = df[c].loc[0]
  return df,sh,em,row
```

```python
model.train(em, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
anger_sub = model.wv.vectors

```

#AFFINITY PROPPAGATION **CLUSTERING**

```python
import numpy as np
from sklearn.cluster import AffinityPropagation

def cluster(a):
    clustering = AffinityPropagation().fit(a)
    labels = clustering.labels_

    n_clusters = len(np.unique(labels))
    cluster_data = []

    for cluster_label in range(n_clusters):
        cluster_points = a[labels == cluster_label]
        cluster_data.append(cluster_points)

    synthetic_data = []

    for cluster_points in cluster_data:
        num_samples = len(cluster_points)
        indices = np.random.choice(cluster_points.shape[0], size=num_samples, replace=True)
        synthetic_points = cluster_points[indices]
        synthetic_data.append(synthetic_points)

    n_clusters = len(np.unique(labels))
    print("Number of clusters:", n_clusters)
    return
```

```python
cluster(anger_sub)
```

# **subemotion**

```python
import numpy as np
from sklearn.cluster import AffinityPropagation
import pickle

def calculate_averages(data):
    clustering = AffinityPropagation().fit(data)
    labels = clustering.labels_

    n_clusters = len(np.unique(labels))
    cluster_averages = []

    for cluster_label in np.unique(labels):
        cluster_examples = data[labels == cluster_label]
        cluster_average = np.mean(cluster_examples, axis=0)
        cluster_averages.append(cluster_average)

    return np.array(cluster_averages)

def save_averages_to_file(averages_variable, output_filename):
    # Save all cluster averages to a .npy file
    np.save(output_filename, averages_variable)
    print("Saved all cluster averages to:", output_filename)

 # Print the dimensions of the saved file
    loaded_data = np.load(output_filename)
    print("Dimensions of saved file:", loaded_data.shape)
    # Print the variable name
    print("Variable name: averages_variable")

    return averages_variable  # Return the variable containing cluster averages

```

```python
anger_sub = model.wv.vectors
output_filename = 'anger.npy'
averages_variable = calculate_averages(anger_sub)
save_averages_to_file(averages_variable, output_filename)
```

```python
import numpy as np

def append_npy_files(file1, file2,file3,file4,file5,file6,file7,file8,file9,file10, output_file):
    data1 = np.load(file1)
    data2 = np.load(file2)
    data3 = np.load(file3)
    data4 = np.load(file4)
    data5 = np.load(file5)
    data6 = np.load(file6)
    data7 = np.load(file7)
    data8 = np.load(file8)
    data9 = np.load(file9)
    data10 = np.load(file10)


    appended_data = np.concatenate((data1, data2,data3,data4,data5,data6,data7,data8,data9,data10), axis=0)

    np.save(output_file, appended_data)

if __name__ == "__main__":
    file1 = "/content/anger.npy"
    file2 = "/content/anticipation.npy"
    file3 = "/content/disgust.npy"
    file4 = "/content/fear.npy"
    file5 = "/content/joy.npy"
    file6 = "/content/negative.npy"
    file7 = "/content/positive.npy"
    file8 = "/content/sadness.npy"
    file9 = "/content/surprise.npy"
    file10 = "/content/trust.npy"
    output_file = "/content/subemotion_orginal.npy"

    append_npy_files(file1, file2,file3,file4,file5,file6,file7,file8,file9,file10, output_file)
    print("Files appended and saved to", output_file)
```
